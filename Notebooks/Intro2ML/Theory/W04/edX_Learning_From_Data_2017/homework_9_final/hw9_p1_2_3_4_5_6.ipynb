## Problem 1

Let's have a look at an example from Homework 5, Problem 3. We had a polynomial transform of order $Q = 4$ and dimension $d=2$:

$\mathbf{x} \mapsto  \mathbf{z} = (1,   x_1,       x_2,     x_1^2,   x_1 x_2, 
  x^2_2, x_1^3, x^2_1 x_2, x_1 x_2^2,     x_2^3,
  x_1^4, x_1^3 x_2, x_1^2 x_2^2, x_1 x_2^3, x_2^4)
  $

The vector $\mathbf{z}$ has $15$ components which we can list as follows:

- constant: $1$
- order $1$: $ x_1,       x_2$
- order $2$: $x_1^2,   x_1 x_2, x^2_2$
- order $3$: $x_1^3, x^2_1 x_2, x_1 x_2^2,     x_2^3$
- order $4$: $x_1^4, x_1^3 x_2, x_1^2 x_2^2, x_1 x_2^3, x_2^4$

Let's count the [monomials](https://www.mathsisfun.com/definitions/monomial.html) in our example:

- constant: $1$, number of monomials $ = 1$
- order $1$: $ x_1,       x_2$, number of monomials $ = 2$
- order $2$: $x_1^2,   x_1 x_2, x^2_2$, number of monomials $ = 3$
- order $3$: $x_1^3, x^2_1 x_2, x_1 x_2^2,     x_2^3$, number of monomials $ = 4$
- order $4$: $x_1^4, x_1^3 x_2, x_1^2 x_2^2, x_1 x_2^3, x_2^4$, number of monomials $ = 5$

You can see that there are $k+1$ monomials of order $k$, because the monomials have the form $x_1^k x_2^0, x_1^{k-1} x_2^1, x_1^{k-2} x_2^2, ..., x_1^{1} x_2^{k-1}, x_1^{0} x_2^k,$ with a total of $k+1$ monomials, e.g. for $k = 2$ we have the $k+1 = 3$ monomials $x_1^2,   x_1 x_2, x^2_2$. Each monomial corresponds to an entry in the $\mathbf{z}$ vector, so the total number of monomials gives us the length of the vector $\mathbf{z}$ which is equal to the dimensionality of the $\cal{Z}$ space.

In total we have $1 + 2 + 3 + 4 + 5 = 15$ monomials, so the dimensionality should be $15$. And indeed, we can confirm this by counting the number of components of $\mathbf{z}$.

In general, if the highest order is $Q$, we have $1 + 2 + ... + (Q+1) = \frac{(Q+1)(Q+2)}{2}$ monomials, so the dimensionality of $\cal{Z}$ is $\frac{(Q+1)(Q+2)}{2}$. In our example we have $Q = 4$ which yields the dimensionality $\frac{(Q+1)(Q+2)}{2} = \frac{(4+1)(4+2)}{2} = 15$.

The problem asks us for the dimensionality of $\cal{Z}$ if the highest order is $Q = 10$. Using this value we get the dimensionality $\frac{(Q+1)(Q+2)}{2} = \frac{(10+1)(10+2)}{2} = 66$. Ignoring the constant coordinate $z_0 = 1$ the answer is $65$, which is not among the given choices, so the correct answer is **1[e]**.

## Problem 2

The problem asks us to compute the average hypothesis $\bar{g}$ and check whether $\bar{g} \in \cal{H}$. This means we have to consider all possible data sets $\cal{D}$ and average over the learned hypotheses $g^{(\cal{D})}$.

$\bar{g}(\mathbf{x}) = \mathbb{E}_{\cal{D}} \left[ g^{(\cal{D})}(\mathbf{x}) \right] $

If we only have a finite number of data sets ${\cal{D}}_1, {\cal{D}}_2,..., {\cal{D}}_K$, then we can compute $\bar{g}$ as:

$\bar{g}(\mathbf{x}) = \frac{1}{K} \sum_{i=1}^{K} g^{({\cal{D}}_i)}(\mathbf{x})$

Let us explore the given options.

________

**2[a]** If $\cal{H}$ consists of a single hypothesis $h$, i.e. ${\cal{H}} = \{ h \}$, then for every data set ${\cal{D}}_i$ the learned hypothesis $g^{({\cal{D}}_i)}$ is the same, namely $g^{({\cal{D}}_i)} = h$. So we have 

$\bar{g}(\mathbf{x}) = \frac{1}{K} \sum_{i=1}^{K} g^{({\cal{D}}_i)}(\mathbf{x}) = \frac{1}{K} \sum_{i=1}^{K} h(\mathbf{x}) = \frac{1}{K} K h(\mathbf{x})  = h(\mathbf{x})$, 

so $\bar{g} = h \in \cal{H}$ which means $\bar{g}$ is in $\cal{H}$. Thus, we can exclude **2[a]** as answer.

______

**2[b]** If $\cal{H}$ is the set of all constant, real-valued hypotheses, i.e. ${\cal{H}} = \{ h \mid h(\mathbf{x}) = const \in \mathbb{R} \}$, then for every data set ${\cal{D}}_i$ the learned hypothesis is $g^{({\cal{D}}_i)}(\mathbf{x}) = const_i$. So we have

$\bar{g}(\mathbf{x}) = \frac{1}{K} \sum_{i=1}^{K} g^{({\cal{D}}_i)}(\mathbf{x}) = \frac{1}{K} \sum_{i=1}^{K} const_i = const$. 

So $\bar{g}$ is a constant function, which means $\bar{g}$ is in $\cal{H}$. Thus, we can exclude answer **2[b]**.

____

**2[c]** If $\cal{H}$ is the linear regression model, i.e. ${\cal{H}} = \{ h \mid h(\mathbf{x}) = \mathbf{w}^T \mathbf{x} \}$, then for every data set ${\cal{D}}_i$ the learned hypothesis is $g^{({\cal{D}}_i)} = (\mathbf{w}^i)^T \mathbf{x}$. So we have 

$\bar{g}(\mathbf{x}) = \frac{1}{K} \sum_{i=1}^{K} g^{({\cal{D}}_i)}(\mathbf{x}) = \frac{1}{K} \sum_{i=1}^{K} (\mathbf{w}^i)^T \mathbf{x}$

$= \frac{1}{K} \left[ (\mathbf{w}^1)^T \mathbf{x} + (\mathbf{w}^2)^T \mathbf{x} + ... + (\mathbf{w}^K)^T \mathbf{x} \right]$

$ = \frac{1}{K} \left[ (w^1_0 x_0 + w^1_1 x_1 + ... + w^1_d x_d) +  (w^2_0 x_0 + w^2_1 x_1 + ... + w^2_d x_d) + ... +  (w^K_0 x_0 + w^K_1 x_1 + ... + w^K_d x_d) \right] $

$ = \frac{1}{K} \left[ (w^1_0 + w^2_0 + ... + w^K_0)x_0 + (w^1_1 + w^2_1 + ... + w^K_1)x_1 + ... + (w^1_d + w^2_d + ... + w^K_d)x_d \right] $

$ = \left[ \frac{1}{K}(w^1_0 + w^2_0 + ... + w^K_0)x_0 + \frac{1}{K}(w^1_1 + w^2_1 + ... + w^K_1)x_1 + ... + \frac{1}{K}(w^1_d + w^2_d + ... + w^K_d)x_d \right] $

$ = \left[ \bar{w}_0x_0 + \bar{w}_1 x_1 + ... + \bar{w}_d x_d \right]$

$ = \bar{\mathbf{w}}^T \mathbf{x} $

This means $\bar{g}(x)$ is in $\cal{H}$, and we can exclude answer **2[c]**.

**Note:**  $\mathbf{w}^i$ does not denote a power but the weight vector associated with the learned hypothesis $g^{({\cal{D}}_i)}$ using the data set ${\cal{D}}_i$.

___

**2[d]** If $\cal{H}$ is the logistic regression model, i.e. ${\cal{H}} = \left\{ h \mid h(\mathbf{x}) = \frac{1}{1 + \exp(-\mathbf{w}^T \mathbf{x})} \right\}$, then for every data set ${\cal{D}}_i$ the learned hypothesis is $\frac{1}{1 + \exp(-(\mathbf{w}^i)^T \mathbf{x})} $. 

So we have

$\bar{g}(\mathbf{x}) = \frac{1}{K} \sum_{i=1}^{K} g^{({\cal{D}}_i)}(\mathbf{x}) =  \frac{1}{K} \sum_{i=1}^{K} \frac{1}{1 + \exp(-(\mathbf{w}^i)^T \mathbf{x})} $

We will now show that there exists $\bar{g} \notin \cal{H}$. Suppose our input space is ${\cal{X}} = [-10,10]$. Further, let's assume that we only have two data sets ${\cal{D}}_1$ and ${\cal{D}}_2$, and that our learning algorithm returns the following learned hypotheses:

$g^{({\cal{D}}_1)}(\mathbf{x}) = \frac{1}{1 + \exp(-(\mathbf{w}^1)^T \mathbf{x})} = \frac{1}{1 + \exp(-(w^1_0 x_0 + w^1_1 x_1))} =  \frac{1}{1 + \exp(-(0.5 x))}$

$g^{({\cal{D}}_2)}(\mathbf{x}) = \frac{1}{1 + \exp(-(\mathbf{w}^2)^T \mathbf{x})} = \frac{1}{1 + \exp(-(w^2_0 x_0 + w^2_1 x_1))} = \frac{1}{1 + \exp(-(0.6 x))}$

___

Let us compute the average $\bar{g}$:

$\bar{g}(\mathbf{x}) = \frac{1}{2} \left[ \frac{1}{1 + \exp(-(0.5 x))} + \frac{1}{1 + \exp(-(0.6 x))}   \right] $

___

Suppose $\bar{g} \in \cal{H}$, then we can express $\bar{g}$ as  

$\bar{g}(\mathbf{x}) = \frac{1}{1 + \exp(-\mathbf{w}^T \mathbf{x})} = \frac{1}{1 + \exp(-(w_0 + w_1 x_1))} $ (with $x_0 = 1$)

___

We can set both expressions equal to obtain **this equation**:

$ \frac{1}{2} \left[ \frac{1}{1 + \exp(-(0.5 x))} + \frac{1}{1 + \exp(-(0.6 x))}\right]  =  \frac{1}{1 + \exp(-(w_0  + w_1 x_1))} $

___

Let's examine this equation for $x_1 = 0$:

$ \frac{1}{2}  \left[ \frac{1}{1 + \exp(-(0.0))} + \frac{1}{1 + \exp(-(0.0))}\right]  =  \frac{1}{1 + \exp(-(w_0))} $

$\Rightarrow  \frac{1}{1 + \exp(-(0.0))} =  \frac{1}{1 + \exp(-(w_0))} $

___

It follows that $w_0 = 0$ which we plug into **the equation**:

$ \frac{1}{2} \left[ \frac{1}{1 + \exp(-(0.5 x))} + \frac{1}{1 + \exp(-(0.6 x))}\right]  =  \frac{1}{1 + \exp(-(w_1 x_1))} $

___

Let's solve for the constant $w_1$:

$ \frac{1}{2} \left[ \frac{1}{1 + \exp(-(0.5 x))} + \frac{1}{1 + \exp(-(0.6 x))}\right]  =  \frac{1}{1 + \exp(-(w_1 x_1))} $

$\Rightarrow  1 + \exp(-(w_1 x_1)) =\left( \frac{1}{2} \left[ \frac{1}{1 + \exp(-(0.2 + 0.5 x))} + \frac{1}{1 + \exp(-(0.6 x))}\right] \right)^{-1} $

$\Rightarrow  \exp(-(w_1 x_1)) =\left( \frac{1}{2} \left[ \frac{1}{1 + \exp(-(0.5 x))} + \frac{1}{1 + \exp(-(0.6 x))}\right] \right)^{-1} - 1$

$\Rightarrow  -(w_1 x_1) = \ln \left\{ \left( \frac{1}{2} \left[ \frac{1}{1 + \exp(-(0.5 x))} + \frac{1}{1 + \exp(-(0.6 x))}\right] \right)^{-1} - 1 \right\}$

$\Rightarrow  w_1 x_1 = -\ln \left\{ \left( \frac{1}{2} \left[ \frac{1}{1 + \exp(-(0.5 x))} + \frac{1}{1 + \exp(-(0.6 x))}\right] \right)^{-1} - 1 \right\}$

$\Rightarrow  w_1 = -\frac{1}{x_1} \left( \ln \left\{ \left( \frac{1}{2} \left[ \frac{1}{1 + \exp(-(0.5 x))} + \frac{1}{1 + \exp(-(0.6 x))}\right] \right)^{-1} - 1 \right\}\right)$

Let's check if $w_{1}$ is indeed a constant by plotting the right hand side.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as style
style.use('bmh')

def theta(s):
    return 1 / (1 + np.exp(-s))

def right_hand_side(x):
    return -(np.log(( 0.5*( theta(0.5*x) + theta(0.6*x) ))**(-1) -1)) / x

def main():
    x = np.arange(-1,1, 0.01) + 0.001
    y = right_hand_side(x)
    
    plt.plot(x,y)
    plt.title("Is $w_1$ a constant?")
    plt.xlabel('$x_1$')
    plt.ylabel("$w_1$ = right hand side\n(according to assumption)")
    plt.show()
    
    
main()
```

We can see that according to the plot $w_1$ is not a constant which contradicts the definition of $w_1$ being constant. Therefore, our assumption that the average $\bar{g}$ with

$\bar{g}(\mathbf{x}) = \frac{1}{2} \left[ \frac{1}{1 + \exp(-(0.5 x))} + \frac{1}{1 + \exp(-(0.6 x))}   \right] $

is in ${\cal{H}} = \left\{ h \mid h(\mathbf{x}) = \frac{1}{1 + \exp(-\mathbf{w}^T \mathbf{x})} \right\}$ is wrong. Thus, answer **2[d]** is correct.

## Problem 3

Option **3[d]** states that we can determine if there is overfitting by comparing the values of the difference $(E_{out} - E_{in}$) only without knowing $E_{in}$ and $E_{out}$.


However, by definition overfitting occurs if $E_{in}$ decreases while $E_{out}$ increases. This means we need to know the values of $E_{in}$ and $E_{out}$ to determine if there is overfitting, and it is not sufficient to know the difference $(E_{out} - E_{in}$) only. 

Thus, the correct answer to Problem 3 is **3[d]**.
