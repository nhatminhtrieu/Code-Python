## Problem 20 - Aggregation


Given two learned hypothesis $g_1$ and $g_2$ we can form the aggregate hypothesis $g$ given by 

$g(\mathbf{x}) = \frac{1}{2}(g_1(\mathbf{x}) + g_2(\mathbf{x}))$ for all $\mathbf{x} \in \cal{X}$.

**Claim**

$E_{\text{out}}(g)$ cannot be worse than the average of $E_{\text{out}}(g_1)$ and $E_{\text{out}}(g_2)$.

___

**Proof**

We can express the claim as follows:

$E_{\text{out}}(g) \leq \frac{1}{2} \left[ E_{\text{out}}(g_1) + E_{\text{out}}(g_2) \right]$

We will now show how to derive that inequality.

**Step 1**. We start with the following true statement:

$0 \leq [g_1(x) - g_2(x)]^2$

___

**Step 2**. Expand:

$\Rightarrow 0 \leq g_1(x)^2 - 2g_1(x)g_2(x) + g_2(x)^2$

___

**Step 3**. We add $2g_1(x)g_2(x)$ to both sides:

$\Rightarrow 2g_1(x)g_2(x) \leq g_1(x)^2 + g_2(x)^2$

___

**Step 4**. We divide both sides by 4:

$\Rightarrow \frac{1}{2}g_1(x)g_2(x) \leq \frac{1}{4} g_1(x)^2 + \frac{1}{4} g_2(x)^2  $

___

**Step 5**. We add $-g_1(x)f(x) -g_2(x)f(x)$ to both sides:

$\Rightarrow  \frac{1}{2}g_1(x)g_2(x)-g_1(x)f(x) -g_2(x)f(x) \leq  \frac{1}{4} g_1(x)^2 + \frac{1}{4} g_2(x)^2 -g_1(x)f(x) -g_2(x)f(x)$

___

**Step 6**. We add  $\frac{1}{4}g_1(x)^2 +  \frac{1}{4}g_2(x)^2  + f(x)^2$  to both sides:


$\Rightarrow \frac{1}{4}g_1(x)^2 + \frac{1}{2}g_1(x)g_2(x) + \frac{1}{4}g_2(x)^2 - 2 \cdot \frac{1}{2}(g_1(x)+g_2(x))f(x) + f(x)^2            \leq \frac{1}{2}g_1(x)^2 - g_1(x)f(x) + \frac{1}{2} g_2(x)^2 - g_2(x)f(x) +  f(x)^2$

___

**Step 7**. We simplify both sides:

$\Rightarrow \left[\frac{1}{2} \{ g_1(\mathbf{x})+g_2(\mathbf{x}) \} -f(\mathbf{x})\right]^2 \leq \frac{1}{2} \left[  \{g_1(\mathbf{x})-f(\mathbf{x})\}^2 + \{g_2(\mathbf{x})-f(\mathbf{x})\}^2\right]$

___

**Step 7**. We form the sum over the points $x_1, x_2, ..., x_N$:

$\Rightarrow \sum_{i=1}^N \left[\frac{1}{2} \{ g_1(\mathbf{x}_i)+g_2(\mathbf{x}_i) \} -f(\mathbf{x}_i)\right]^2 \leq \frac{1}{2} \left[ \sum_{i=1}^N \{g_1(\mathbf{x}_i)-f(\mathbf{x}_i)\}^2 + \sum_{i=1}^N \{g_2(\mathbf{x}_i)-f(\mathbf{x}_i)\}^2\right]$

___

**Step 8**. On the left hand side we replace $\frac{1}{2} \{ g_1(\mathbf{x}_i)+g_2(\mathbf{x}_i) \}$ by $g(\mathbf{x}_i)$:

$\Rightarrow \sum_{i=1}^N \left[g(\mathbf{x}_i) -f(\mathbf{x}_i)\right]^2 \leq \frac{1}{2} \left[ \sum_{i=1}^N \{g_1(\mathbf{x}_i)-f(\mathbf{x}_i)\}^2 + \sum_{i=1}^N \{g_2(\mathbf{x}_i)-f(\mathbf{x}_i)\}^2\right]$

___

**Step 9**. We divide both sides by $N$:

$\Rightarrow \frac{1}{N}\sum_{i=1}^N \left[g(\mathbf{x}_i) -f(\mathbf{x}_i)\right]^2 \leq \frac{1}{2} \left[ \frac{1}{N}\sum_{i=1}^N \{g_1(\mathbf{x}_i)-f(\mathbf{x}_i)\}^2 + \frac{1}{N}\sum_{i=1}^N \{g_2(\mathbf{x}_i)-f(\mathbf{x}_i)\}^2\right]$

___

**Step 10**. We use the definition of $E_{\text{out}}$ as mean squared error:

$\Rightarrow E_{\text{out}}(g) \leq \frac{1}{2} \left[ E_{\text{out}}(g_1) + E_{\text{out}}(g_2) \right]$

This is the inequality that we wanted to derive which completes our proof.
