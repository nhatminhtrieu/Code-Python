## Problem 19 - Bayesian Priors

For this problem I recommend reading the lecture notes of [Introduction to Probability and Statistics (MIT Course Number
18.05)](https://ocw.mit.edu/courses/mathematics/18-05-introduction-to-probability-and-statistics-spring-2014/index.htm), in particular chapter 12 and 13 on *Bayesian updating with discrete and continuous priors*.

Bayes's rule says the following:

$P(\cal{H} \mid \cal{D}) = \frac{P(\cal{D} \mid \cal{H} ) P(\cal{H})}{P(\cal{D})} $

with 

- $P(\cal{H} \mid \cal{D})$: posterior
- $P(\cal{D} \mid \cal{H})$: likelihood
- $P(\cal{H})$: prior 
- $P(\cal{D} \mid \cal{H} ) P(\cal{H})$: Bayes numerator

We can see that 

$P(\cal{H} \mid \cal{D}) \propto P(\cal{D} \mid \cal{H} ) P(\cal{H})$

i.e.

$\text{posterior} \propto \text{likelihood} \times \text{prior}$
___

We are given the following:

- $f \in [0,1]$ : unknown probability of getting a heart attack
- $h \in [0,1]$ : hypothesis to model $f$
- prior $p(h = f)$: probability density that is uniform over $h \in [0,1]$
- Data (sample point): states whether a randomly chosen person from the population had suffered from a heart attack


We have to compute the posterior. The posterior in general is computed as follows.

$\text{posterior} \propto \text{likelihood} \times \text{prior}$

If $x$ denotes the data (sample point), then $x = 1$ means we observed a heart attack. We can then write the following:

- posterior: $p(h=f \mid x = 1)$
- likelihood: $P(x = 1 \mid h=f)$
- prior: $p(h=f)$

So we get

$\text{posterior} \propto \text{likelihood} \times \text{prior}$

$\Rightarrow p(h=f \mid x = 1) \propto P(x = 1 \mid h=f) \times p(h=f)$

Note that with capital $P$ we describe a probability, whereas small $p$ denotes a probability density. Thus, our posterior $p(h=f \mid x = 1)$ is a probability density, and $p(h \in [f+df] \mid x = 1) df$ is the probability that our hypothesis is in the interval $[f+df]$ given the data $x = 1$.

See also [this discussion](https://math.stackexchange.com/questions/273120/notation-for-probability-density) on the notation for a probability density.
___

The likelihood $P(x = 1 \mid h=f)$ is the probability that a person had had a heart attack according to the given model $h=f$, but we know that this is described by the probability value $f$, so we can simply write $P(x = 1 \mid h=f) = f$

___

The likelihood $P(x = 0 \mid h=f)$ is the probability that a person has not suffered from a heart attack for a given model $h=f$, and this is described by the probability value $(1-f)$, so we can simply write $P(x = 0 \mid h=f) = (1-f)$

___

The prior $p(h = f)$ is uniform over $h \in [0,1]$, so we have $p(h = f) = 1$ (constant probability density over the interval $[0,1]$. This is also known as a *flat* prior. Thus we have

$p(h=f \mid x = 1) \propto P(x = 1 \mid h=f) \times p(h=f)$

$\Rightarrow p(h=f \mid x = 1) \propto f \times 1$

___

We can see that the posterior $p(h=f \mid x = 1)$ increases linearly with $f$. Thus, **19[b]** is the correct answer.
