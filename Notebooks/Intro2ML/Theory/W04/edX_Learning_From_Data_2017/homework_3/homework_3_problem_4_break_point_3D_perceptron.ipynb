# Problem 4

We can approach the problem as follows:

- For $N = 4$ points the maximum number of dichotomies that can be generated by all possible 2D perceptron lines is $14$, i.e. $m_{H}(4) = m_{2D}(4) = 14$ . This is less than $16 = 2^N$, so $N = 4$ is a break point (see slides 9/20 and 17/20 of the lecture 5 notes).

- A configuration of $N = 4$ points that generates only 14 dichotomies for the 2D perceptron is shown on slide 9/20 of the lecture 5 notes. 

- However, this configuration can be separated by the 3D perceptron model if we raise a diagonal pair of points into the z-direction, see the illustration below. So the 3D perceptron model can generate the maximum of $16 = 2^N$ dichotomies for $N=4$ points, and we conclude that $N = 4$ is *not* a break point for the 3D perceptron model. We say that the 3D perceptron model shatters $N = 4$ points: 

$$m_{H}(4) = m_{3D}(4) = 16 = 2^4 = 2^N$$


- The next step would be to check if we can find $N = 5$ points that can be shattered by the 3D perceptron model. However, doing so would require us to check $2^N = 32$ possible dichotomies. And on top of that we would have to do this in three dimensions which is a little difficult.

- Even worse though is that if we wanted to prove that $N = 5$ is a break point, then we would have to check every possible set of $N = 5$ points, and for each of them show that the number of possible dichotomies is less than $2^N$ to conclude  $m_{H}(5) = m_{3D}(5) < 2^5$ . This is impossible since we have infinitely many  sets of 5 points. Alternatively we could find a "general" constellation of $N$ points that is representative for all possible sets of $N$ points like in problem 6. Maybe such a representative constellation would be 5 points on a sphere.

- Instead I made an educated guess. For the 2D perceptron hypothesis set the break point is $k=4$ which equals (dimension + 2). So I guessed that for the 3D perceptron the break point would be (dimension + 2) = (3 + 2) = 5. So I marked **[d]5** as answer for problem 4.
