# Problem 6

In this problem our hypothesis set $\cal{H}$ is given by two intervals on the real line. The points are classified as follows:

- If a point is within an interval, it is assigned the value $+1$, 
- otherwise the point is assigned the value $-1$.

What is the smallest break point?

---------------

First, we find the largest number of N points that can still be shattered by our hypothesis (the largest number of N points for which our hypothesis can generate $2^N$ dichotomies), see the illustration below. 


![2 Intervals hypothesis set shatters N = 4 points](homework_3_figures/homework_3_problem_6_hypothesis_2_intervals_shatters_4_points.png)


Second, we show that for every possible set of $N = 5$ points it is not possible to generate $32 = 2^N$ dichotomies.
We consider the particular dichotomy below which has alternating colors. 

It is important to note that this particular set of 5 points is representative for all possible set of 5 points, e.g. if we take another set of 5 points whose distances to each other are different compared to the ones depicted below, then it doesn't change the number of possible dichotomies since we can stretch the intervals arbitrarily. In that sense all sets of 5 points are equivalent to each other.

![2 Intervals hypothesis set cannot shatter N = 5 points](homework_3_figures/homework_3_problem_6_hypothesis_2_intervals_cannot_shatter_5_points.png)

We recognize that we cannot generate that dichotomy with only two intervals. Therefore, for $N = 5$ our 2-Intervals hypothesis set cannot generate all $2^N$ possible dichotomies. This means the 2-Intervals hypothesis set cannot shatter 5 points, and we conclude that $N = 5$ is the smallest break point. So answer **6[c]** is correct.

# Problem 7

What is the growth function for our "2-Intervals" hypothesis set?

We start by considering the dichotomies that can be generated by using only a single interval out of the two intervals.
There are $N+1$ possible spots where we can put the two ends of the interval. This gives us ${{N + 1} \choose 2}$ possible dichotomies, see also slide 12/20 of the lecture 5 notes.

Using both intervals there are $N+1$ possible spots where we can put the four ends of both intervals without them overlapping (if the intervals overlap then it is the same as if we had only a single interval). This gives us ${{N + 1} \choose 4}$ possible dichotomies.

And finally we have to account for the case where both ends of an interval lie in the same spot such that they don't contain any points. This corresponds to all points being classified as $-1$ and in the image to all points being red. There is only one such dichotomy, so we have to add 1.

In total we get

$$ m_{\cal{H}}(N) = {{N + 1} \choose 2} + {{N + 1} \choose 4} + 1  $$

So answer **7[c]** is correct.

# Problem 8

What is the (smallest) break point for the "M-Intervals" learning model?

We have already seen in Problem 6 that the dichotomy with three blue points that are separated by two red points cannot be generated by the "2-Intervals" hypothesis set. The reason is that we can "cover" the first two blue points with the two available intervals, but we are missing an additional interval to cover the third blue point.

The same applies to the "M-Intervals" hypothesis set. See the illustration below:

![M Intervals hypothesis set cannot shatter N = 2M+1 points](homework_3_figures/homework_3_problem_8_hypothesis_M_intervals_cannot_shatter_2M_plus_1_points.png)

We see that this particular dichotomy of $N = 2M+1$ points cannot be generated by the "M-Intervals" hypothesis set. This means that we cannot generate all $2^N$ possible dichotomies, and we conclude that $N = 2M + 1$ is a break point.

However, is this the smallest break point?

- a) We know that for the "$M = 1$ interval hypothesis set" the smallest break point is 3 (see slide 18/20 of lecture 5).
- b) From problem 6 we know that for the "$M = 2$ intervals hypothesis set" the smallest break point is 5.

Let us check if these numbers satisfy the formula $N = 2M + 1$:
- a) For $M = 1$ the formula predicts a break point of $N = 2M + 1 = 2\cdot 1 + 1 = 3$ which is indeed the case.
- b) For $M = 2$ the formula predicts a break point of $N = 2M + 1 = 2\cdot 2 + 1 = 5$ which is indeed the case.

So we can conclude that the smallest break point is $N = 2M + 1$, and answer **8[d]** is correct.
