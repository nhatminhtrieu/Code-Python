# Problem 9

In this problem we are given a triangle. If a point in the plane lies within the triangle, it is classified as +1, else as -1. Our hypothesis set consists of all triangles in the plane.

What is the largest number of points that can be shattered by this hypothesis set?

-----

We first show that $N = 7$ can be shattered by the triangle hypothesis set. We choose our $N$ points to be on a circle, see also lecture slide 13/20 of lecture 5. We then argue that all $2^N$ dichotomies can be generated by considering the case where 0, 1, 2, ... 7 points lie within the triangle.

For each case we will examine only a few examples, since looking at all examples would require $2^N = 2^7 = 128$ sketches.

![triangles_shatter_7_points](homework_3_figures/homework_3_problem_9_triangles_shatter_7_points.png)

Second, we show that $N = 9$ points cannot be shattered by the triangle hypothesis set. Again, we will not examine all examples but instead have a look at one example.

![triangles_cannot_shatter_9_points](homework_3_figures/homework_3_problem_9_triangles_cannot_shatter_9_points.png)

Assuming that this configuration of $N = 9$ points on the circle can generate the most dichotomies among all 9-point configurations, i.e. there is no other set of 9 points that can generate more dichotomies, we can conclude that the triangle hypothesis set cannot shatter 9 points. (It's important to note that we did not present a rigorous proof here since we only assumed that the points on the circle is the 'best' configuration that generates the most dichotomies.)

We conclude that $N = 7$ is the largest number of points that can be shattered by the triangle hypothesis set.

The correct answer is therefore **[d] 7** .
