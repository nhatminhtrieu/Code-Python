## Problem 2

We are given a plot of the box $[-1,1] \times [-1,1]$ with a hyperbola separating the box into three regions. Points in the left and right region are classified as $-1$, the points in the center are classified as $+1$. 

We have to choose the weight vector $\tilde{w}$ in the transformed space from the given multiple choices such that the given classification is achieved.

```python
import numpy as np
import matplotlib.pyplot as plt


def classify(X, w):
    return np.sign(np.dot(X, w))


def problem2():
    
    w_tilde = np.array([0.2, -2, 1])
    N = 1000
    
    x1 = np.random.uniform(-1,1,N)
    x2 = np.random.uniform(-1,1,N)
    Z = np.array([np.ones(N), x1*x1, x2*x2]).T   # feature matrix
    
    classification = classify(Z, w_tilde)
    
    plt.plot(x1[classification==-1], x2[classification==-1], 'r_')
    plt.plot(x1[classification==+1], x2[classification==+1], 'b+')
    
    plt.ylim(-1,1)
    plt.xlim(-1,1)
    plt.show()



problem2()
```

The plot above was generated by choosing the weight vector $\tilde{w} = [0.2, -2, 1]$ , so we have $\tilde{w}_1 < 0$ and $\tilde{w}_2 > 0$. The answer to this problem is therefore **2[d]** .
