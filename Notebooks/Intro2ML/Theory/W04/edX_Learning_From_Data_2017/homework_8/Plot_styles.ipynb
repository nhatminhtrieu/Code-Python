```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

import matplotlib.style as style
print(style.available)

# style.use('fivethirtyeight')
# style.use('seaborn')
style.use('bmh')
```

```python
N = 10
x = np.arange(N)
y_noise = x + 2 * np.random.uniform(-1,1,N)
y = x

fig1 = plt.figure(1, figsize = (12,8), dpi = 300)

plt.plot(x, y, 'bo-', label='no noise')
plt.plot(x, y_noise,'ro-', label='with noise')

plt.title('Line vs noisy line')
plt.xlabel('month')
plt.ylabel('temperature')
plt.legend()
plt.show()
```
