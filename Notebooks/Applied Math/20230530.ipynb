```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
x = np.linspace(0, 1)
plt.plot(x, np.exp(x**2))
plt.axis("equal")
plt.show()
```

```python
def integral(N):
    dx = 1/N
    x = np.linspace(0, 1, N + 1)
    x_star = (x[1:] + x[:-1])/2
    return dx*np.sum(np.exp(x_star**2))
```

```python
integral(100)
```

```python
integral(150)
```

```python
def poly_appr(N):
    x = np.linspace(0, 1, N)
    X = np.c_[x**0]
    for i in range(1, N):
        X = np.c_[X, x**i]
    y = np.c_[np.exp(x**2)]
    return (np.linalg.inv(X) @ y).reshape((-1,))

def eval(poly, x):
    y = 0
    for i in range(len(poly)):
        y += poly[i]*x**i
    return y
```

```python
x = np.linspace(0, 1, 100)
plt.plot(x, np.exp(x**2))
plt.plot(x, eval(poly_appr(4), x))
plt.show()
```

```python
poly = poly_appr(4)
S = np.sum(poly/np.arange(1, len(poly) + 1))
S
```

```python
A = sp.Matrix([
    [4, -14, 1, -53],
    [8, -10, 1, -41],
    [-8, 6, 1, -25]
])
A
```

```python
A.echelon_form()
```

```python
A.rref()
```

```python
a, b, c = sp.symbols("a b c")
sp.linsolve(A, (a, b, c))
```

```python
a = 7/8
b = -11/4
c = 1
```

```python
x = np.linspace(-1, 5, 100)
y = a*x**2 + b*x + c
plt.plot(x, y)
plt.axis("equal")
plt.scatter([0, 2, 4], [1, -1, 4])
plt.show()
```

$$
