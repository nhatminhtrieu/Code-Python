```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Load image
img = Image.open('cat.jpg')

# Convert image to numpy array
img = np.array(img)

kmeans = KMeans(n_clusters=3, random_state=0, n_init='auto', max_iter=100).fit(img.reshape(-1, 3))

# Get cluster labels
labels = kmeans.labels_

# Get cluster centers
centers = kmeans.cluster_centers_

# Get cluster centers for each pixel
img_compressed = centers[labels].reshape(img.shape).astype('uint8')

# Show image
Image.save(Image.fromarray(img_compressed), 'cat.')
```
